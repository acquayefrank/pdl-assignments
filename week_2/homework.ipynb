If you've skipped the url above out of sloth, here's the game of life:
* You have a 2D grid of cells, where each cell is "alive"(1) or "dead"(0)
* Any living cell that has 2 or 3 neighbors survives, else it dies [0,1 or 4+ neighbors]
* Any cell with exactly 3 neighbors becomes alive (if it was dead)

For this task, you are given a reference numpy implementation that you must convert to pytorch.
numpy code inspired by: https://github.com/rougier/numpy-100


__Note:__ You can find convolution in `torch.nn.functional.conv2d(Z,filters)`. Note that it has a different input format. 


````python
from scipy.signal import correlate2d as conv2d

def np_update(Z):
    # Count neighbours with convolution
    filters = np.array([[1,1,1],
                        [1,0,1],
                        [1,1,1]])
    
    N = conv2d(Z,filters,mode='same')
    
    # Apply rules
    birth = (N==3) & (Z==0)
    survive = ((N==2) | (N==3)) & (Z==1)
    
    Z[:] = birth | survive
    return Z
````

````python
def torch_update(Z):
    """
    Implement an update function that does to Z exactly the same as np_update.
    :param Z: torch.FloatTensor of shape [height,width] containing 0s(dead) an 1s(alive)
    :returns: torch.FloatTensor Z after updates.
    
    You can opt to create new tensor or change Z inplace.
    """
    
    filters = torch.tensor([[1,1,1],
                            [1,0,1],
                            [1,1,1]], dtype=torch.float)
    new_data = F.pad(input=Z, pad=(1, 1, 1, 1), mode='constant', value=0)
    N = F.conv2d(new_data.unsqueeze(0).unsqueeze(0), filters.unsqueeze(0).unsqueeze(0))
    
     # Apply rules
    birth = (N==3) & (Z==0)
    survive = ((N==2) | (N==3)) & (Z==1)
    
    Z[:] = birth | survive
    return Z
````

````python
#initial frame
Z_numpy = np.random.choice([0,1],p=(0.5,0.5),size=(100,100))
Z = torch.from_numpy(Z_numpy).type(torch.FloatTensor)

#your debug polygon :)
Z_new = torch_update(Z.clone())

#tests
Z_reference = np_update(Z_numpy.copy())
assert np.all(Z_new.numpy() == Z_reference), "your pytorch implementation doesn't match np_update. Look into Z and np_update(ZZ) to investigate."
print("Well done!")
````

````python
from notmnist import load_notmnist
X_train, y_train, X_test, y_test = load_notmnist(letters='ABCDEFGHIJ')
X_train, X_test = X_train.reshape([-1, 784]), X_test.reshape([-1, 784])
````

````python
%matplotlib inline
plt.figure(figsize=[12,4])
for i in range(20):
    plt.subplot(2,10,i+1)
    plt.imshow(X_train[i].reshape([28,28]))
    plt.title(str(y_train[i]))
````

````python
from torch.utils.data import DataLoader

batch_size = 128

train_data, test_data = [], []

for image, label in zip(X_train, y_train):
    train_data.append([image, label])
    
for image, label in zip(X_test, y_test):
    test_data.append([image, label])

trainloader = DataLoader(train_data, batch_size=batch_size)
testloader = DataLoader(test_data, batch_size=batch_size)
````

````python
model = nn.Sequential(
     nn.Linear(784, 1024),
     nn.Softmax(),
     nn.Linear(1024, 10)
)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
loss_func = nn.CrossEntropyLoss()
````

````python
def train(model, train_loader, optimizer):
    train_loss, train_acc = [], []
    for train_data in train_loader:
        X_train, y_train = train_data
        X_train = X_train.type(torch.FloatTensor)
        y_train = y_train.type(torch.LongTensor)

        y_predicted = model(X_train)
        loss = loss_func(y_predicted, y_train)
        
        max_index = torch.max(y_predicted, 1,keepdim=True)[1]
        target_matches = torch.eq(max_index, y_train.data.view(len(y_train), 1))
        accuracy = torch.sum(target_matches).numpy() / len(y_train)
        
        train_loss.append(loss.item())
        train_acc.append(accuracy)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    return np.mean(train_loss), np.mean(train_acc)
````

````python
def validate(model, val_loader):
    val_loss, val_acc = [], []
    for test_data in val_loader:
        X_test, y_test = test_data
        X_test = X_test.type(torch.FloatTensor)
        y_test = y_test.type(torch.LongTensor)
        
        y_predicted = model(X_test)
        loss = loss_func(y_predicted, y_test)
        
        max_index = torch.max(y_predicted, 1,keepdim=True)[1]
        target_matches = torch.eq(max_index, y_test.data.view(len(y_test), 1))
        accuracy = torch.sum(target_matches).numpy() / len(y_test)
        
        val_loss.append(loss.data.item())
        val_acc.append(accuracy)        

    
    return np.mean(val_loss), np.mean(val_acc)
````

````python
from IPython.display import clear_output


for epoch in range(30):
    train_loss, train_accuracy  = train(model, trainloader, optimizer)
    validation_loss, validation_accuracy = validate(model, testloader)
    clear_output()
    print(f"Epoch : {epoch}")
    print(f"Train loss: {train_loss}")
    print(f"Train accuracy: {train_accuracy}")
    print(f"Validation loss: {validation_loss}")
    print(f"Validation accuracy: {validation_accuracy}")
````
